# Data with Pandas

```python
import pandas as pd
```

```python
file_path = "../data/eistage.csv"

df = pd.read_csv(
    file_path,
    sep=";",
    encoding="utf-8",
    low_memory=False,
    skiprows=[0, 1],
    usecols=[3,5,6],
    nrows=70,
    names=["year", "heat_days", "ice_days"],
    parse_dates=["year"]
)
```
